# Task 2:Build Advanced Preprocessing Function

```python
import re
def preprocess_text(text):
    
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)
    
    # 3. Remove numbers (using loop)
    clean_text = ""
    for char in text:
        if not char.isdigit():
            clean_text += char
    
    # 4. Handle repeated characters (soooo → so)
    temp = ""
    prev_char = ""
    for char in clean_text:
        if char != prev_char:
            temp += char
        prev_char = char
    clean_text = temp
    
    # 5. Remove punctuation
    temp = ""
    for char in clean_text:
        if char.isalnum() or char == " ":
            temp += char
    clean_text = temp
    
    # 6. Remove extra spaces
    clean_text = " ".join(clean_text.split())
    
    # 7. Remove very short words (≤2) except "no", "not"
    final_words = []
    for word in clean_text.split():
        if len(word) > 2 or word in ["no", "not"]:
            final_words.append(word)
    
    clean_text = " ".join(final_words)
    
    return clean_text


# Test
print(preprocess_text("WOW!!! This is soooo goooood!!! Visit http://example.com now, I have 2 dogs"))
```

# Task 4: Token Analytics

```python
def token_analytics(sentences):
    results = []
    for sent in sentences:
        tokens = sent.split()
        
        total_tokens = len(tokens)
        unique_tokens = len(set(tokens))
        # average token length
        if total_tokens > 0:
            avg_length = sum(len(word) for word in tokens) / total_tokens
        else:
            avg_length = 0
        
        results.append({
            "sentence": sent,
            "total_tokens": total_tokens,
            "unique_tokens": unique_tokens,
            "avg_token_length": round(avg_length, 2)
        })
    
    return results
# Example processed sentences
sentences = [
    "wow this sooo good visit now have dogs",
    "this product not good",
    "free offer click now"
]
# Run
output = token_analytics(sentences)

for res in output:
    print(res)
```

#Task 5: Frequency Analysis

```python
from collections import Counter

def frequency_analysis(sentences):
    
    all_tokens = []
    
    # Combine all tokens
    for sent in sentences:
        tokens = sent.split()
        all_tokens.extend(tokens)
    
    # Count frequency
    freq = Counter(all_tokens)
    
    # Top 10 most frequent
    most_common = freq.most_common(10)
    
    # Top 5 least frequent
    least_common = freq.most_common()[:-6:-1]
    
    return most_common, least_common


# Example processed sentences
sentences = [
    "wow this sooo good visit now have dogs",
    "this product not good",
    "free offer click now"
]

# Run
top10, least5 = frequency_analysis(sentences)

print("Top 10 most frequent words:")
print(top10)

print("\nTop 5 least frequent words:")
print(least5)
```

#Task 6: Build Full Pipeline

```python
import re  # used for regex (pattern matching like URLs, emails)

def preprocess_text(text):
    
    # 🔹 Convert text to lowercase (WOW → wow)
    text = text.lower()
    
    # 🔹 Remove URLs and email patterns
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)
    
    # 🔹 Remove numbers using loop
    temp = ""
    for char in text:
        if not char.isdigit():   # keep only non-digit characters
            temp += char
    text = temp
    
    # 🔹 Handle repeated characters (allow max 2 times)
    temp = ""
    count = 1
    for i in range(len(text)):
        if i > 0 and text[i] == text[i-1]:  # same as previous character
            count += 1
            if count <= 2:   # allow only 2 repetitions
                temp += text[i]
        else:
            count = 1
            temp += text[i]
    text = temp
    
    # 🔹 Remove punctuation (keep only letters, numbers, spaces)
    temp = ""
    for char in text:
        if char.isalnum() or char == " ":
            temp += char
    text = temp
    
    # 🔹 Remove extra spaces (multiple → single space)
    text = " ".join(text.split())
    
    # 🔹 Remove short words (length ≤ 2) except "no", "not"
    words = []
    for word in text.split():
        if len(word) > 2 or word in ["no", "not"]:
            words.append(word)
    
    # 🔹 Return cleaned sentence
    return " ".join(words)


def full_pipeline(text_list):
    
    clean_sentences = []  # store cleaned sentences
    all_tokens = []       # store all tokens (words)
    
    for text in text_list:
        clean_text = preprocess_text(text)  # clean each sentence
        clean_sentences.append(clean_text)
        
        tokens = clean_text.split()  # convert sentence to words
        all_tokens.extend(tokens)    # add to global token list
    
    # 🔹 Return final output as dictionary
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# ✅ Test input
texts = [
    "WOW!!! This is soooo goooood!!! Visit http://example.com now",
    "I have 2 dogs!!!",
    "Freeeee offer!!! Click now!!!"
]

# 🔹 Run pipeline
result = full_pipeline(texts)

# 🔹 Print output
print(result)
```

```python
import re

def preprocess_text(text):
    
    # 🔹 Handle empty input
    if not text or not isinstance(text, str):
        return ""
    
    # lowercase
    text = text.lower()
    
    # remove URLs & emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)
    
    # remove numbers
    temp = ""
    for char in text:
        if not char.isdigit():
            temp += char
    text = temp
    
    # handle repeated characters (max 2)
    temp = ""
    count = 1
    for i in range(len(text)):
        if i > 0 and text[i] == text[i-1]:
            count += 1
            if count <= 2:
                temp += text[i]
        else:
            count = 1
            temp += text[i]
    text = temp
    
    # remove punctuation & emojis (keep only letters/numbers/space)
    temp = ""
    for char in text:
        if char.isalnum() or char == " ":
            temp += char
    text = temp
    
    # remove extra spaces
    text = " ".join(text.split())
    
    # remove short words (≤2) except no, not
    words = []
    for word in text.split():
        if len(word) > 2 or word in ["no", "not"]:
            words.append(word)
    
    return " ".join(words)


def full_pipeline(text_list):
    
    clean_sentences = []
    all_tokens = []
    
    for text in text_list:
        clean_text = preprocess_text(text)
        
        # 🔹 Skip empty results safely
        if clean_text:
            clean_sentences.append(clean_text)
            tokens = clean_text.split()
            all_tokens.extend(tokens)
        else:
            clean_sentences.append("")  # keep position safe
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# ✅ Test cases (Error Handling)
texts = [
    "",                    # empty
    "😂😂🔥🔥",             # only emojis
    "1234567890",          # only numbers
    "WOW!!! 123 sooo 😂 good!!!"
]

result = full_pipeline(texts)

print(result)
```
